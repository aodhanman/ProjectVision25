```python
import pandas as pd
import numpy as np

from sklearn.discriminant_analysis import LinearDiscriminantAnalysis, QuadraticDiscriminantAnalysis
from sklearn.metrics import confusion_matrix, classification_report, precision_score

%matplotlib inline
```

```python
df = pd.read_csv('Smarket.csv', usecols=range(1,9), index_col=0, parse_dates=True)
df.head()
```

```python
import pandas as pd

# Re-load the Smarket.csv to include the 'Year' column for splitting
df_full = pd.read_csv('Smarket.csv')

# Use the 'Year' column to split the data into training and testing sets
X_train = df_full[df_full['Year'] <= 2004][['Lag1','Lag2']]
y_train = df_full[df_full['Year'] <= 2004]['Direction']

X_test = df_full[df_full['Year'] > 2004][['Lag1','Lag2']]
y_test = df_full[df_full['Year'] > 2004]['Direction']

lda = LinearDiscriminantAnalysis()
model = lda.fit(X_train, y_train)

print(model.priors_)
```

```python
print(model.means_)
```

```python
print(model.coef_)
```

```python
pred=model.predict(X_test)
print(np.unique(pred, return_counts=True))
```

```python
print(confusion_matrix(pred, y_test))
print(classification_report(y_test, pred, digits=3))
```

```python
qda = QuadraticDiscriminantAnalysis()
model2 = qda.fit(X_train, y_train)
print(model2.priors_)
print(model2.means_)
```

```python
pred2=model2.predict(X_test)
print(np.unique(pred2, return_counts=True))
print(confusion_matrix(pred2, y_test))
print(classification_report(y_test, pred2, digits=3))
```
